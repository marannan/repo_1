```python
import numpy as np
import pandas as pd

cols = ["EIN", "NAME", "SUBSECCD", "ACTIV1", "ACTIV2", "ACTIV3", "GRRECMEM",
        "GRRECOTH", "ASS_EOY", "NETINC", "TOTREV", "ZIP", "zip5"]

df = pd.read_csv('../data/core_other/core_other_2013.txt',
                 usecols=cols, low_memory=False)

sec_ids = [1, 12, 14, 15, 16, 50, 60, 80]

orgs = df[df.SUBSECCD.isin(sec_ids)]

zip_codes = pd.read_csv('../data/zipcodes/zip_codes.csv')

merged = orgs.merge(zip_codes, left_on="zip5", right_on="Zipcode")

pd.__version__
```

```python
grouped = merged[["SUBSECCD", "ASS_EOY", "TOTREV", "NETINC"]].groupby(['SUBSECCD'])
x = grouped.agg(['size','mean', 'median', 'sum'])
x.to_excel('../results/table1.xlsx')
```
